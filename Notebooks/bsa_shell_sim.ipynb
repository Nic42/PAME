# 12/20/12 BSA shell filling simulation


In understanding the influence of large particle aggregates on optical fibers, I thought that by simulating BSA on various sized particles, I could get some understanding on the relative change in signal intensity that we'd expect.  I hypothesize that larger particles will be less sensitive to a BSA layer because of the shell is much smaller than the particle itself.  

```python
from simparser import SimParser #Interface for handling simulation results
from pyuvvis import divby, absplot, specplot, boxcar
from IPython.core.display import Image
from pandas import Series
```

Using simparser, I can load the runresults for the simulation directly from disk.  I did three simulations for np's of sizes d=20nm, 50nm, 200nm respectively

```python
s20=SimParser('20nm.mdat')
s50=SimParser('50nm.mdat')
s200=SimParser('200nm.mdat')
```

We can see that this simulated incremented only one variable, the particle shell fill fraction from 0-26 (about 0-50% coverage):

```python
print s20.df_parms.index
print s20.df_parms.values  #Shows ( iterated over the volume fraction of the shell material from 0-0.26
```

I can also print out all of the other static parameters (eg that were not changed during the simulation) with a simple method, **write_parms**.  This also has the option to output to a file.

```python
s20.write_parms(full=True)
```

All of the simulation result arrays are stored in s.results, which is a panel that can pull up any array in the program.  For example, I can pull up the extinction spectra of the nanoparticles particles as the shell mixing was incremented using the **'ext'** key as easily as:

```python
s20.results['ext'].plot(title='Extinction spectra', legend=False)
```

It makes more sense to look at the relative change, which I can easily get using **divby()** from pyuvvis.  We can do this for all 3 sets of results.

```python
ext20=divby(s20.results['ext'])
ext50=divby(s50.results['ext'])
ext200=divby(s200.results['ext'])
x=Series(ext20.index)

plt.subplots(3,2, figsize=(16,20))
subplot(322).plot(x, ext20)
subplot(321).plot(x, s20.results['ext'])
subplot(324).plot(x, ext50)
subplot(323).plot(x, s50.results['ext'])
subplot(326).plot(x, ext200)
subplot(325).plot(x, s200.results['ext'])

subplot(321).set_title('20nm Np Reflectance')
subplot(322).set_title('20nm Np RELATIVE Reflectance')
subplot(323).set_title('20nm Np Reflectance')
subplot(324).set_title('20nm Np RELATIVE Reflectance')
subplot(325).set_title('20nm Np Reflectance')
subplot(326).set_title('20nm Np RELATIVE Reflectance')
```

Note: change in the 200nm particle's extinction spectrum is so small that the plot axis is actually reading 1.000XYZ.  See below

# Corresponding Fiber results

I can look at the resuts on an optical fiber, **with a few caveats**.

1. I am not using optimized equivalent particles in my mixing routine. (aka have not used composite equiv mixing method)
2. Because I'm changing particle size, the fiber coverage is different between these systems, so if the response is coverage dependent, this will impact results.
3. This only looks at the TM modes of the fiber, not the TE modes.

```python
### Get the reflectance and it's relative version
R20=s20.results['Ravg']
R20abs=divby(R20)

R50=s50.results['Ravg']
R50abs=divby(R50)

R200=s200.results['Ravg']
R200abs=divby(R200)

plt.subplots(3,2, figsize=(16,20))
subplot(321).plot(x, R20)
subplot(321).set_title('20nm Np Reflectance')
subplot(322).plot(x, R20abs)
subplot(322).set_title('20nm Np RELATIVE Reflectance')
subplot(323).plot(x, R50)
subplot(323).set_title('50nm Np Reflectance')
subplot(324).plot(x, R50abs)
subplot(324).set_title('50nm Np RELATIVE Reflectance')
subplot(325).plot(x, R200)
subplot(325).set_title('200nm Np Reflectance')
subplot(326).plot(x, R200abs)
subplot(326).set_title('200nm Np RELATIVE Reflectance')



```
