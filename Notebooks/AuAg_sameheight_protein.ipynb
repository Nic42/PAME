```python
from pame.simparser import LayerSimParser #Interface for handling simulation results
from skspec import *
from skspec.plotting import splot
from skspec.data import aunps_glass
```

## Parameters for PAME

1. AuNP: d = 80nm, Vfrac = 0.2386, sites = 218444, disk cov = 45.5692%.

2. AgNP: d = 60nm, Vfrac = 0.0976, sites = 158854.

2. Using LinearSum model for the mixture of gold and silver, alpha = beta = 0.5.

3. Protein layer thickness: 6nm (neutravidin). Linker layer thickness: 2.5nm (PAH).

4. Simulation over protein shell Vfrac: 0.0-0.5, 10 steps. In other words for AuNPs, 0-617 binding sites out of 646 are taken, shell coverage: 0-95.51%. As for AgNPs, 0-363 binding sites out of 381 are taken, shell coverage: 0-95.28%.

## Protein Binding Response

```python
ax1, ax2, ax3, ax4 = splot(2,2, figsize=(12,8))

s1=LayerSimParser.load_pickle('../Simulations/AuAg_sameheight_protein1.mpickle')
s1.backend='skspec'

ref1=s1.primary_panel()['R_avg'][350:700]
ref1.plot(ax=ax1, title='Protein binding on AuNP')

ref1.reference = 0
ref1.plot(ax=ax3, title='', norm='r')

s2=LayerSimParser.load_pickle('../Simulations/AuAg_sameheight_protein2.mpickle')
s2.backend='skspec'

ref2=s2.primary_panel()['R_avg'][350:700]
ref2.plot(ax=ax2, title='Protein binding on AgNP')

ref2.reference = 0
ref2.plot(ax=ax4, title='', norm='r')
```

```python
ax1, ax2, ax3, ax4 = splot(2,2, figsize=(12,8))

slice1 = ref1.ix[460:700]
slice1.plot(ax=ax1, title='Protein binding on AuNP')
slice1.plot(ax=ax3, title='', norm='r')

slice2 = ref2.ix[365:460]
slice2.plot(ax=ax2, title='Protein binding on AgNP')
slice2.plot(ax=ax4, title='', norm='r')
```
